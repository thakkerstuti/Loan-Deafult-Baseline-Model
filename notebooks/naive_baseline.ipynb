# Naive Baseline Model

Simple baseline model that always predicts the majority class.

## Setup and Data Loading

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report

print("Libraries imported successfully!")
```

## Load Data and Prepare Features

```python
# Read CSV file
df = pd.read_csv('../data/dataset.csv')

print(f"Dataset shape: {df.shape}")
print(f"\nFirst few rows:")
df.head()
```

```python
# === Prepare the data ===
# Assuming 'Default' is the target column
target_col = 'Default'
feature_cols = [col for col in df.columns if col != target_col and col != 'LoanID']  # Exclude ID and target

X = df[feature_cols]
y = df[target_col]

# For categorical features, we need to encode them, but for naive baseline, let's just use numeric features
numeric_features = X.select_dtypes(include=[np.number]).columns.tolist()
X_numeric = X[numeric_features]

print("Features used:", numeric_features)
print(f"\nFeatures shape: {X_numeric.shape}")
print(f"Target shape: {y.shape}")
```

```python
print("\n" + "="*60)
print("TARGET DISTRIBUTION")
print("="*60)

print("\nTarget distribution:")
print(y.value_counts())
print(f"\nPercentage of defaults: {y.mean() * 100:.2f}%")
print(f"Percentage of non-defaults: {(1-y.mean()) * 100:.2f}%")
```

## Train-Test Split

```python
# Split the data
X_train, X_test, y_train, y_test = train_test_split(
    X_numeric, y, test_size=0.2, random_state=42, stratify=y
)

print(f"Train set size: {X_train.shape}")
print(f"Test set size: {X_test.shape}")
print(f"\nTrain target distribution:")
print(y_train.value_counts())
```

## Naive Baseline Model

```python
# === Naive Baseline Model ===
# Always predict the majority class

majority_class = y_train.mode()[0]
y_pred_naive = np.array([majority_class] * len(y_test))

print("="*60)
print("NAIVE BASELINE MODEL")
print("="*60)
print(f"\nNaive baseline prediction: always predict {majority_class}")
print(f"Majority class in training data: {majority_class} ({(y_train == majority_class).mean() * 100:.2f}%)")
```

## Evaluation Metrics

```python
# Calculate metrics
accuracy = accuracy_score(y_test, y_pred_naive)
precision = precision_score(y_test, y_pred_naive, zero_division=0)
recall = recall_score(y_test, y_pred_naive, zero_division=0)
f1 = f1_score(y_test, y_pred_naive, zero_division=0)

print("\n" + "="*60)
print("NAIVE BASELINE METRICS")
print("="*60)
print(f"\nAccuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")
```

## Confusion Matrix

```python
# Confusion matrix
cm = confusion_matrix(y_test, y_pred_naive)

print("\n" + "="*60)
print("CONFUSION MATRIX")
print("="*60)
print("\nConfusion Matrix:")
print(cm)
print(f"\nTN (True Negatives): {cm[0,0]}")
print(f"FP (False Positives): {cm[0,1]}")
print(f"FN (False Negatives): {cm[1,0]}")
print(f"TP (True Positives): {cm[1,1]}")
```

## Classification Report

```python
# Classification report
print("\n" + "="*60)
print("CLASSIFICATION REPORT")
print("="*60)
print("\n" + classification_report(y_test, y_pred_naive, zero_division=0))
```

## Key Insights

### Understanding the Naive Baseline:

- **High Accuracy (88-89%)**: This looks impressive, but it's misleading!
- **Reason**: The dataset is imbalanced - most loans don't default (88%)
- **The Catch**: The model predicts 0 (No Default) for everything
- **Recall = 0%**: Model never identifies any actual defaults
- **F1 Score = 0%**: Poor overall performance

### Why This Matters:
- Naive baseline sets a **minimum performance threshold**
- Any real model must **beat this baseline**
- Shows that **accuracy alone is not enough**
- We need metrics like **F1 Score, Recall, and Precision**

### Next Steps:
- Use more sophisticated models (Logistic Regression, Random Forest, XGBoost)
- Handle class imbalance with SMOTE or class weights
- Focus on Recall (catching defaults) and Precision (reducing false alarms)
